# **Splitting the dataset**


```python
df.drop('Id',axis=1,inplace=True)   #dropping the ID column as it is not helpful to our dataset 
df.head()
```

```python
X=df[["SepalLengthCm",  "SepalWidthCm", "PetalLengthCm",  "PetalWidthCm"]]   #Selecting columns
print(X.shape)
y=df[["Species"]]                                                  #Selecting target variable. They can also be selected by converting them to dummy variables and then predciting
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)

clf = DecisionTreeClassifier(random_state=0)
clf.fit(X_train,y_train) 
print("Training Complete")
```

```python
from sklearn import metrics
y_hat = clf.predict(X_test)
print('The accuracy of the Decision Tree is',metrics.accuracy_score(y_hat,y_test))
```

# **Plotting the Tree**

```python
from sklearn.datasets import load_iris
from sklearn import tree

clf1 = tree.DecisionTreeClassifier(random_state=0)
iris = load_iris()
plt.figure(figsize=(17,17))
clf1 = clf1.fit(iris.data, iris.target)

tree.plot_tree(clf1,filled=True, fontsize=15) 
plt.show()
```
